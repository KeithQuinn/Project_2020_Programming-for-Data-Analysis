## 2 Extrusion Process Overview

The extrusion process begins at the hopper of the machine. The polymer material in granulated form is loaded into the hopper and gravity fed onto the screw. The screw is situated within a barrel that is heated to a controlled temperature. The screw rotates and conveys the material along the barrel. As the material is conveyed it is heated and transformed from a granular state to a molten state. As the molten material leaves the barrel it is delivered through an assembly of tooling to an extrusion head. The purpose of the extrusion head is to provide a passage for the molten polymer to flow and to house the cone and die.  The cone and die creates the required shape of the product [1]. Refer to Figure 1 which is a cross section of the extruder, this shows the hopper, screw, heaters and barrel etc which have just been discussed.



When the material exits from the extrusion head it is passed through a water bath for cooling. The extruded material is pulled at a constant rate from the extrusion head by a puller. The puller feeds the tube to a cutter which cuts the product to the required length. Before getting cut to length the material passes through a measurement system which is constantly measuring the required process outputs. The product is then transferred from the extrusion line via a conveyor [1]. Figure 2 provides an image of the extrusion line complete with downstream ancillary equipment.

## 3 Process Validation Overview

Process validation is defined as the collection and evaluation of data, from the process design stage through commercial production, which establishes scientific evidence that a process is capable of consistently delivering quality product. Process validation involves a series of activities taking place over the lifecycle of the product and process [7].

For this project process validation is split into three stages and a brief explanation of each is provided;

- Product Design
- Process Development
- Process Validation Builds

### 3.1 Product Design 
At the product design stage, the research and development (R&D) team design the product. This involves designing and manufacturing components of predefined dimensions. These components go through rigorous testing to prove that the design is acceptable from a quality, safety and design for manufacturing standpoints . 

As part of testing, challenge builds are complete to challenge and test the limits of the design. At the end of product design, the part drawing which includes all the specifications is approved, released and handed over to the process development team. ### 3.2 Process Development
At the process development (PD) the manufacturing process is developed. The process inputs are identified and their relationship to the process outputs studied. ### 4.2.1 Process Inputs

*__Barrel Temperature [C]__* 

It critical that the barrel temperatures are suitable for the process, different processes with require different temperatures, for example a polyethylene extrusion process requires less thermal energy to break intermolecular bonds than a polyamide extrusion process. If the temperature is too low the screw may not be able to rotate in the barrel, this could cause the extruder to alarm out in a best case scenario or snap the screw in a worst case scenario. If the temperature is too high the material degrades which impacts its performance. Given the wide range of possible extrusion processes, temperatures can vary from 100 to 350 degrees celcius. The temperature is controlled using proportional–integral–derivative (PID) controllers and the inputs are floating point numbers with two decimal places.

See code that follows for the generation of a random barrel temperature input, also included is a randomly generated list of 5 temperatures defined as temperature profile. Most extruders require a temperature profile, that is different zones on the barrel require different temperature inputs, in the example below a profile having 5 zones is considered.

```python
# Simulated data point for barrel temperature [C]
# Let's assume an extrusion process for PEBAX and the range is 200 - 240 [C]

import random
Barrel_Temp = round(random.uniform(200, 240),2) # round to 2 decimal places
print(f'Barrel Temperature is: {Barrel_Temp} Degrees Celcius')
print()
Barrel_Profile = [] # empty list to append to
for i in range(0,5): # for loop
    n = round(random.uniform(200, 240),2) # generates a random 
    Barrel_Profile.append(n) # appends to the list values
print(f'Barrel Profile is: {Barrel_Profile} Degrees Celcius')
```

*__Screw speed [RPM]__*

The screw speed is the roational speed the screw rotates within the extruder barrel. As the screw rotates it conveys material forward through the barrel. This conveying of material controls the amount of material output, the faster the roation the higher the output and vice versa.

Just like barrel speed, screw speed is dependent on the type of extrusion process, type of material, extruded tibe size etc. For example if there is a large tube manufactured on a small extruder (relatively speaking) the screw would be running very fast i.e. 100 rpm, but if there's a small tube manufactured on a large extruder (again relatively speaking) the screw would be running very slow i.e. 10 rpm. 

If the screw rotates too slow the extrusion line can pulse which is an inconsistent output, the result of this would be an inconsistent tube that would fail inspection for outside diameter. If the screw rotates too fast it can over shear the polymer material, this can lead to chain scission which impacts the mechanical properties of the material.

Screw speed is controlled on through an inbuilt controller on the extruder and the input is a non-zero integer. See code that follows for the generation of a random screw speed input value.

```python
# Simulated data point for screw speed [rpm]
# Let's assume an average size extruder and tube 40-60 [rpm]

screw_speed = random.randint(40, 60)
print(f'The screw speed is: {screw_speed} [rpm]')
```

*__Puller speed [mm/s]__*

The puller speed is the speed at which the extrusion is pulled through the cooling tank, OD gauge and into the cutter, refer to Figure 2. The puller speed is important as it controls how the polymer chains are orientated as the enter the cooling tank. The orientation of the polymer chains is an important factor when considering mechanical properties. For example if the chains are fully aligned forming a high percentage crystallinity there may be high strength but low elongation tensile properties, the tube may be brittle. If on the other hand the chains have time ro relax before being entering the cooling tank then the tube will exhibit completely different mechanical properties such as ductility and toughness.

The puller speed works closely with the screw speed often in closed loop control, this is because of the important relationshhip between the two. If the screw speed increased and the puller speed remained constant then the OD of the tube will increase, also is the screw speed stayed constant and the puller speed increased the OD of the tube would decrease. The process works on a fine balance of keeping the two in synchronisation in order to manufacture a consistent tube.

As with barrel temperature and screw speed, puller speed is dependent on the process, thick walled extrusions would be pulled slower than thin walled at the same screw speed to get the desired output. If toughness and ductility is important the extrusion would be given time to relax before entering the cooling tank so this would require a slow puller speed. If the tube is to be hard wearing then it would be best to have the chains aligned to give a highly crystalline structure, this would be better achieved with a fast puller speed. The following code generates a random non zero integer as would be representative of the puller speed input.

```python
# Simulated data point for puller [mm/s]
# Let's assume a puller speed window of of 180-220 [mm/s]

puller_speed = random.randint(180, 220)
print(f'The puller speed is: {puller_speed} [mm/s]')
```

### 4.2.2 Process Outputs

*__Outer Diameter [mm]__*

The outer diameter is the diameter of the outside of the tube. As the tube passes through the laser gauge (refer to Figure 2) the OD is measured. It records continuous variable data having two decimal places. The data follows a normal distribution. An example dataset is generated in the following code. the data set assumes the nominal value from the drawing as the mean and assumes a standard deviation of 0.1 and 1000 values.

*__Normal Distribution__*

This is the first time normal distribution has been mentioned. A brief description of a normal distribution is provided.

The normal distribution is the most important probability distribution in statistics because it fits many natural phenomena. For example, heights, blood pressure, measurement error, and IQ scores follow the normal distribution. It is also known as the Gaussian distribution and the bell curve.

The normal distribution is a probability function that describes how the values of a variable are distributed. It is a symmetric distribution where most of the observations cluster around the central peak and the probabilities for values further away from the mean taper off equally in both directions. Extreme values in both tails of the distribution are similarly unlikely [12].

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

loc = 1.732 # mean
scale = 0.1 # standard deviation of 0.1 is assumed
size = 1000 # number of values

normal_data = np.random.normal(loc, scale, size) # generate a data set that follows a normal distribution

plt.figure(figsize=(10, 6)) # figure size
plt.title('Normal Distribution') # title
plt.suptitle('Extruded Tube Outer Diameter') # subtitle
plt.xlabel(f'Mean = {loc}mm, Std Dev = {scale}mm and total values = {size}') # xlabel
plt.ylabel('Occurrences') # ylabel
sns.distplot(a=normal_data, hist=True, kde=True, color='g')
txt="Figure 6: Tube OD" # create caption
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10) # locate caption
plt.show() # show histogram
```

*__Wall Thickness [mm]__*

The wall thickness of the tube is measured again as the tube passes through the laser gauge (refer to Figure 2) the OD and the ID are measured. The gauge is connected to a data processor which calculates the wall thickness based on the OD and ID. The formula is simply $ (OD-ID) / 2 $.

Wall thickness is a 2 sided specification having continuous variable data to two decimal places. It follows a normal distribution. An example dataset is generated in the following code. The data set assumes the nominal value from the part drawing in Figure 4 as the mean and assumes a standard deviation of 0.1 and 1000 values.

```python
loc = 0.187 # mean
scale = 0.1 # standard deviation
size = 1000 # number of values

normal_data = np.random.normal(loc, scale, size) # generate a data set that follows a normal distribution

plt.figure(figsize=(10, 6)) # figure size
plt.title('Normal Distribution') # title
plt.suptitle('Extruded Tube Wall Thickness') # subtitle
plt.xlabel(f'Mean = {loc}mm, Std Dev = {scale}mm and total values = {size}') # xlabel
plt.ylabel('Occurrences') # ylabel
sns.distplot(a=normal_data, hist=True, kde=True, color='r')
txt="Figure 7: Wall thickness" # create caption
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10) # locate caption
plt.show() # show histogram
```

*__Tube Length [mm]__*

The length of the tube is measured using a ruler that has 1mm increments. The data set consists of integer values and the specification is one sided per the part drawing in Figure 4. The data follows a Gumbel Distribution as a result of the one sided specification. 

*__Gumbel Distribution__*

This is the first time Gumbel distribution has been mentioned. A brief description of the Gumbel distribution is provided.

Gumbel distribution otherwise known as the extreme type I distribution has two forms. One is based on the smallest extreme and the other is based on the largest extreme. These are called the minimum and maximum cases, respectively. It is an asymmetric, extreme value distribution, used to model maximums and minimums. For example, the EVD Type I has been used to predict earthquakes, floods and other natural disasters, as well as modeling operational risk in risk management and the life of products that quickly wear out after a certain age [13]. Given that the length data has a one sided specification the result is right skewed that fits the Gumbel distribution.

```python
loc = 210 # mean
scale = 5 # standard deviation
size = 1000 # number of values

length_data = np.random.gumbel(loc, scale, size).astype(int)
plt.figure(figsize=(10, 6)) # figure size
plt.title('Gumbel Distribution') # title
plt.suptitle('Extruded Tube Length') # subtitle
plt.xlabel(f'Mean = {loc}mm, Std Dev = {scale}mm and total values = {size}') # xlabel
plt.ylabel('Occurrences') # ylabel
sns.distplot(a=length_data, hist=True, kde=True, color='r')
txt="Figure 7: Length" # create caption
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10) # locate caption
plt.show() # show histogram
```

*__Visual Inspection__*

The visual inspection is completed using a vision system and involves a machine visually checking the parts for any defects such as a scratch or a tear. The results are pass or fail which generates an attribute dataset. The visual inspection is sampled in subgroups of 10 and in the example below 100 subgroups are considered with a probability of 80% chance of pass. The data follows a binomial distribution. A sample dataset is generated in the following code.

*__Binomial Distribution__*

This is the first time binomial distribution has been mentioned. A brief description of binomail distribution is provided.

A binomial distribution can be thought of as simply the probability of a SUCCESS or FAILURE outcome in an experiment or survey that is repeated multiple times. The binomial is a type of distribution that has two possible outcomes (the prefix “bi” means two, or twice). For example, a coin toss has only two possible outcomes: heads or tails and taking a test could have two possible outcomes: pass or fail. If you purchase a lottery ticket, you’re either going to win money, or you aren’t. Basically, anything you can think of that can only be a success or a failure can be represented by a binomial distribution [14]

```python
n = 10 # subgroup size
p = 0.80 # probability of heads or tails
size = 100 # number of subgroups

# Pass is True or 1
# Fail is False or 0

visual_data = np.random.binomial(n, p, size) # create a data set that fits a binomial distribution

plt.figure(figsize=(10, 6)) # figure size
sns.set(style="darkgrid") # set style
ax = sns.countplot(x=visual_data) # plot countplot
plt.title("Binomial Distribution") # title
plt.suptitle("Visual Inspection of Extruded Tube") # subtitle
plt.ylabel("") # ylabel
plt.xlabel(f'Number of Passes from {n} trials and {size} repeats') # xlabel
txt="Figure 8: Binomial Distribution." # create caption
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10) # locate caption
plt.show() # show countplot
```

## 4.4 Process Validation

Now that the process is developed the next step is to validate the process. In this section of the report simulated data is generated representative of process validation and in line with the process development section above.

The process capability (Cpk) is also calculated for the variable outputs; OD, wall thickness and length. This calculated process capability is then compared to the requirements to pass the process validation, this scientific evidence is used to make a decision as to whether the process validation passes or fails.

__*Process Capability (Cp and Cpk)*__

Process capability is the repeatability and consistency of a manufacturing process relative to the customer requirements in terms of specification limits of a product parameter. This measure is used to objectively measure the degree to which your process is or is not meeting the requirements [15].

Capability indices have been developed to graphically portray that measure. Capability indices let you place the distribution of your process in relation to the product specification limits. Capability indices should be used to determine whether the process, given its natural variation, is capable of meeting established specifications. It is also a measure of the manufacturability of the product with the given processes [15].

Process capability can be expressed with an index. Assuming that the mean of the process is centered on the target value, the process capability index Cp can be used. Cp is a simple process capability index that relates the allowable spread of the spec limits (spec range or the difference between the upper spec limit, USL, and the lower specification limit, LSL) to the measure of the actual, or natural, variation of the process, represented by 6 sigma, where sigma is the estimated process standard deviation [15].

Cpk measures not only the process variation with respect to allowable specifications, it also considers the location of the process average.

Cpk is taken as the smaller of either Cpl or Cpu where

Cpl = (X -LSL) / 3 sigma where X is the process mean

Cpu = (USL – X) / 3 sigma where X is the process mean

Many companies are establishing specific process capability targets. They may typically start with 1.33 for supplier qualification and have an expected goal of 2.0. If the process is near normal and in statistical control, Cpk can be used to estimate the expected percent of defective material [15]. 

For this simulated process validation the assumption is that a Cpk value of 2.0 and above qualifies the process, anything less than a Cpk of 2.0 and the process fails validation.

### 4.4.1 Process Validation - Extruded Tube Outer Diameter Simulated Dataset

```python
ProVal_OD_loc = 1.730 # mean
ProVal_OD_scale = 0.0035 # standard deviation of 0.02 from initial run
ProVal_OD_size = 1000 # number of values
OD_USL=1.757
OD_LSL=1.707

OD_Cpu = round((OD_USL - ProVal_OD_loc) / (3*ProVal_OD_scale),2)
OD_Cpl = round((ProVal_OD_loc - OD_LSL) / (3*ProVal_OD_scale),2)
OD_Cpk = np.min([OD_Cpu, OD_Cpl])

print(f' The upper capability index (Cpu) of the tube OD is {OD_Cpu}')
print(f' The lower capability index (Cpl) of the tube OD is {OD_Cpl}')
print(f' The lower of the two indices is the actual process capability (Cpk) of the tube OD has a value of {OD_Cpk}')
print('----------------------------------------------------------------------------------------')

ProVal_OD_normal_data = np.random.normal(ProVal_OD_loc, ProVal_OD_scale, ProVal_OD_size) # generate a data set that follows a normal distribution

plt.figure(figsize=(10, 6)) # figure size
plt.title('Process Validation - Normal Distribution') # title
plt.suptitle('Process Validation - Extruded Tube Outer Diameter') # subtitle
plt.axvline(x=OD_LSL)
plt.axvline(x=1.732)
plt.axvline(x=OD_USL)
plt.xlabel(f'Mean = {ProVal_OD_loc}mm, Std Dev = {ProVal_OD_scale}mm and total values = {ProVal_OD_size}') # xlabel
plt.ylabel('Occurrences') # ylabel
sns.distplot(a=ProVal_OD_normal_data, hist=True, kde=True, color='g')
txt="Figure 17: Process Validation - Tube OD" # create caption
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10) # locate caption
plt.show() # show histogram
```

From Figure 17 above it can be seen that the process ran slightly off of centre and followed a normal distribution. The capability indices were calculated which quantifies how far off centre the process ran. The upper capability index was 2.57 and the lower was 2.19 indicative of the process running slightly off centre towards the LSL. The actual process capability is the lower of the two. Considering the requirement to pass the process validation was a Cpk of 2.0 or greater this process has passed validation and is ready to manufacture commercial product.
