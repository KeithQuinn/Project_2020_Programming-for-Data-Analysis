### Part 1 Process Validation

Process validation is defined as the collection and evaluation of data, from the process design stage through commercial production, which establishes scientific evidence that a process is capable of consistently delivering quality product. Process validation involves a series of activities taking place over the lifecycle of the product and process.

For this project process validation is split into three stages and a brief explanation of each is provided;

- Product Design
- Process Development
- Process Validation

#### 1.1 Product Design 
At the product design stage, the research and development (R&D) team design the product. This involves designing and manufacturing components of predefined dimensions. These components go through rigorous testing to prove that the design is acceptable from a quality, safety and design for manufacturing standpoints. 

As part of testing, challenge builds are complete to challenge and test the limits of the design. At the end of product design, the part drawing which includes all the specifications is approved, released and handed over to the process development team. Refer to Figure 5 for a part drawing of an extruded tube.



__*Figure 1: extruded tube drawing*__

#### 1.2 Process Development
At the process development (PD) the manufacturing process is developed. The process inputs are identified and their relationship to the process outputs studied. Process windows are developed for each input while ensuring that the components can be manufactured within the drawing specification. For example, if the speed is found to be nominal 50[m/s] and a process window developed of +/-10[m/s] then the process can run anywhere between 40 – 60[m/s] while making acceptable components.


__*Figure 2: process window*__

#### 1.3 Process Validation
At the process validation (PV) stage the process is validated. That is the process is tested through various component builds using settings from the developed input process window. It’s important here that standard production variation is included in the builds, examples are, different shifts, different operators and a tooling change. The performance of the process outputs relative to the print outputs is analysed. If the process outputs are tight relative to the specifications the process is highly capable and passes validation. This is scientific evidence that a process is capable of consistently delivering quality product.

### Part 2 Process Overview - Extrusion
The extrusion process begins at the hopper end of the machine. The material (resin) is loaded into the hopper and gravity fed onto the screw. The screw is situated within a barrel that is heated to a controlled temperature. The screw rotates and conveys the resin along the barrel. As the resin is conveyed it is heated and transformed from a granular state to a molten state. As the molten resin leaves the barrel it is delivered through an assembly of tooling to an extrusion head. The purpose of the extrusion head is to provide a passage for the molten polymer to flow and to house the cone and die.  The cone and die creates the required shape of the product [1]. Refer to Figure 1 which is a cross section of the extruder, this shows the hopper, screw, heaters and barrel etc which have just been discussed.

When the material exits from the extrusion head it is passed through a water bath for cooling. The extruded material is pulled at a constant rate from the extrusion head by a puller. The puller feeds the tube to a cutter which cuts the product to the required length. Before getting cut to length the material passes through a measurement system which is constantly measuring the required process outputs. The product is then transferred from the extrusion line via a conveyor [1]. Figure 2 provides an image of the extrusion line complete with downstream ancillary equipment for reference.

### 1.2 Inputs, Process & Outputs

With any manufacuring there are the inputs, the process and the outputs to be considered, the extrusion process is no exception. For this project there will be three process inputs and four process outputs considered, they are;

__Process Inputs:__

- Line speed [m/s] controls the speed at which the extrusion is pulled through the puller.
- Temperature [C] controls the temperature profile of the extruder from the hopper to the cone and die.
- Melt pump speed [rpm] controls the amount of polymer material is pumped through the cone and die.

__Process:__

- Medical device extrusion

__Process Outputs:__

- Outer diameter - the outside diameter of the tube
- Wall thickness - the thickness of the wall of the extrusion
- Length - the length of the tube
- Tactile and Visual - tactile where the tube feels good, no lumps or bumps and visual where it looks good, no imperfections.

Another important consideration is the component specifications. Having the process outputs is one thing but the specifications are required to provide the size and tolerance to which the component must be manufactured, without these there's no way of knowing what size the component needs to be, we must define how long a piece of string must be!

### Part 2 - Process Development

Now that we have a good understanding of the process the next step is to try to run the extrusion line and make the part. For this project let's consider manufacturing the tube from Figure 5. The dimensions for OD, wall thickness and length will be used.

In reality the approach here would be to set up the extrusion line and run. For the initial runs the obective would be to target nominal and at the end of the run complete the analysis of the data to check what distribution the data follows. We can simulate that using python as follows.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

#--------------------------------------------------Plot of OD-------------------------------------------------

loc = 1.732 # mean
scale = 0.05 # standard deviation
size = 1000 # number of values

normal_data = np.random.normal(loc, scale, size) # generate a data set that follows a normal distribution

plt.figure(figsize=(10, 6)) # figure size
plt.title('Normal Distribution') # title
plt.xlabel(f'Mean = {loc}mm, Std Dev = {scale}mm and total values = {size}') # xlabel
plt.ylabel('Occurrences') # ylabel
sns.distplot(a=normal_data, hist=True, kde=True, color='g')
txt="Figure 6: Tube OD" # create caption
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10) # locate caption
plt.show() # show histogram

#-------------------------------------------Plot of Wall thickness---------------------------------------------

loc = 0.187 # mean
scale = 0.1 # standard deviation
size = 1000 # number of values

normal_data = np.random.normal(loc, scale, size) # generate a data set that follows a normal distribution

plt.figure(figsize=(10, 6)) # figure size
plt.title('Normal Distribution') # title
plt.xlabel(f'Mean = {loc}mm, Std Dev = {scale}mm and total values = {size}') # xlabel
plt.ylabel('Occurrences') # ylabel
sns.distplot(a=normal_data, hist=True, kde=True, color='r')
txt="Figure 7: Wall thickness" # create caption
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=10) # locate caption
plt.show() # show histogram
```
